# 시계열 데이터 간단 실습 (KOSPI 데이터를 활용한 시계열 데이터 시각화)

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("./data/kospi.csv")

print(data)

# Data 컬럼을 datetime 형식으로 변환하여 인덱스로 설정
data['Data'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

#그래프그리기
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'])
plt.title('KOSPI Index')
plt.xlabel('Year')
plt.ylabel('Index')
plt.show()
```
