# 트랜드 실습

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

## 시계열 데이터 시각화

```python
#시계열 데이터 생성
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = [10, 12, 11, 15, 9, 13, 8, 14, 10, 16] * 10

#데이터프레임 생성
data = pd.DataFrame({
    'Date' : dates,
    'Value' : values
})


print(data)
```

### 주기 설정, 주기성 분해

```python
period = 7 #이는 데이터가 7일 간격으로 반복되는 패턴을 가지고 있다는 가정!!

#주기성 분해
result = seasonal_decompose(data['Value'], model='additive', period=period)
```

```python
# 추세 계산
trend = result.trend


# 시계열 데이터 시각화
plt.plot(data['Date'], data['Value'], label='Original')
plt.plot(data['Date'], trend, label='Trend')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Trend Extraction in Time Series Data')
plt.legend()
plt.xticks(rotation=-45)
plt.show()
```

Q.10개 데이터를 10번씩 반복했었는데, 10개 데이터 반복인데도 7일 주기랑 맞는 이유?
A.만약 이 10개의 데이터가 주기적인 패턴을 가지고 있다면, 7일 주기와 맞을 수 있습니다. 예를 들어, 데이터가 주중과 주말을 반복하거나 계절적인 변동을 반영하고 있다면, 7일 주기에 맞을 수 있습니다.
