# 상관 분석

- Seaborn 내장 데이터셋인 'tips' 데이터를 활용하여, 식당에서 팁을 받은 금액과 식사 금액 사이의 상관 관계를 분석

```python
import seaborn as sns
import pandas as pd 

tips_data = sns.load_dataset('tips')
print(tips_data)
```

- 'total_bill'과 'tip' 변수 간의 상관 계수를 계산

```python
corr = tips_data[['total_bill', 'tip']].corr()
print(corr)
```

- Seaborn의 'regplot()' 함수를 활용하여, 'total_bill'과 'tip' 변수 간의 산점도와 회귀선을 함께 시각화

```python
sns.regplot(x='total_bill', y='tip', data=tips_data)
```

# 상관계수

- 상관 계수는 두 변수 간의 관계를 수량화하기 위한 지표 중 하나이다. 대표적으로 '피어슨 상관 계수' 가 있다.

- 피어슨 상관 계수는 -1부터 1까지의 범위를 가지며, 두 변수 간의 관계를 수량화하여 표현한다. 
- 1에 가까울수록 강한 양의 상관 관계가 있고, -1에 가까울수록 강한 음의 상관 관계가 있으며, 0에 가까울수록 두 변수 간의
상관 관계가 약하다.

# 피어슨 상관계수 간단 실습

```python
corr = tips_data['total_bill'].corr(tips_data['tip'], method='pearson')
print(corr)
```

# 공분산 (Covariance)

- 공분산(Covariance)은 두 변수 사이의 상관 관계의 방향과 강도를 나타내는 지표 중 하나이다.
- 두 변수가 함께 변화하는 정도를 나타낸다.
- 두 변수가 같은 방향으로 변화하면 양수의 값을, 반대 방향으로 변화하면 음수의 값을 가진다.
- 공분산 값이 0이라면 두 변수는 서로 독립적인 관계를 가진다.

```python
# 공분산 실습 
covariance = tips_data['total_bill'].cov(tips_data['tip'])
print(covariance)
```

# 산점도 (Scatter plot)

- 산점도(scatter plot)란 두 변수 간의 관계를 시각적으로 나타내는 그래프이다.
- 일반적으로 x축에는 독립 변수(independent variable), y축에는 종속 변수(dependent variable)를 놓는다.

```python
# 산점도 실습 
sns.scatterplot(x='total_bill', y='tip', data=tips_data)
```
