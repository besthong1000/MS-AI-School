## 랜덤 데이터셋을 사용한 인자 분석 연습

```python
import numpy as np
from sklearn.decomposition import FactorAnalysis
```

```python
#데이터 생성
n_samples = 100 #샘플
n_feature = 5 #변수 수 

#데이터셋을 랜덤하게 설정
x = np.random.randn(n_samples, n_feature)
print(x)
```

```python
# 인자 분석 수행
n_compoents = 2 #추출한 인자수
factorAnalysis = FactorAnalysis(n_components=n_compoents, random_state=0)
x_factor = factorAnalysis.fit_transform(x)

# 결과 출력
print("원본 데이터셋 크기", x.shape)
print(f"인자 분석 결과 인자 계수 : {n_compoents}")
print(x_factor)
```
