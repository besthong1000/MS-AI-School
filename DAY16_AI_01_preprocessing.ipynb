```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt

import os
from os.path import join
import warnings
warnings.filterwarnings('ignore')
```

먼저 머신러닝의 대표적인 데이터 셋 중 하나인 전복 데이터를 불러오겠습니다. 전복 데이터셋은 수컷, 암컷, 유아기 3개의 범주로 이루어진 범주형 변수와 길이, 직경, 높이, 무게 등 여러 수치형 변수로 이루어져 있습니다. 데이터를 불러온 후 입력으로 사용할 변수들과 레이블로 사용할 성별 변수로 나누겠습니다.



```python
abalone_path = join('.','abalone.txt')
column_path = join('.','abalone_attributes.txt')

abalone_columns = list()
for l in open(column_path):
  abalone_columns.append(l.strip())
#print(abalone_columns)


# Pandas를 이용해서 abalone 데이터를 읽어온다.
data = pd.read_csv(abalone_path, header=None, names=abalone_columns)
data
```

```python
# head(), shape 를 사용해서 데이터를 살펴본다. 
data.head()
```

```python
label= data['Sex']  #뒤에서 label 사용하려고 빼놓음
del data['Sex']
data.head
```

```python
# Scaling (MinMax Scaling 0~1 사이 숫자로 바꿔줌. 특정 숫자가 널뛰기 하지 않도록 사용)
#라이브러리에 있음.
data2 = (data - np.min(data)) / (np.max(data)-np.min(data))
data2
```

```python
# 1) 데이터 전처리 모듈의 불러오기 및 정의하기
from sklearn.preprocessing import MinMaxScaler

mMscaler = MinMaxScaler()
mMscaler.fit(data)
data3 = mMscaler.transform(data)
data3   #head는 pandas 명령어. head 안쓴다.
```

```python
mMscaler = MinMaxScaler()
data4 = mMscaler.fit_transform(data)  #fit(데이터 붙여넣기)+transform(변환)
```

```python
mMscaled_data = pd.DataFrame(data4, columns=data.columns)
mMscaled_data.head()
```

```python
# 2. Standard Scaling (평균=0,편차가 0을 기준으로 발생함.
#가장 높은 값 =1, 가장 낮은 값 =-1,(표준편차 = 1)
#그런데 -1 넘는 값도 있다. <- 이건 뒤에가서 설명함.)
from sklearn.preprocessing import StandardScaler

sdscaler = StandardScaler()
sdscaled_data = sdscaler.fit_transform(data)

sdscaled_data = pd.DataFrame(sdscaled_data, columns=data.columns)
sdscaled_data.head()
```

```python
#Sampling  (종류2가지: oversampling(적은 클래스의 데이터수 증가), undersampling(많은 클래스의 데이터수 감소))
#클래스의 불균형 해소를 위해 각 클래스별 균형을 맞추는 작업
#클래스의 불균형 : 클래스라벨의 비율이 균형X 한쪽으로 치우친 경우

#1. Random Over,Under Sampling
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

ros = RandomOverSampler()
rus = RandomUnderSampler()
```

```python
#Over Sampling
oversampled_data, oversampled_label = ros.fit_resample(data, label)
oversampled_data = pd.DataFrame(oversampled_data, columns=data.columns)

# Under Sampling
undersampled_data, undersampled_label = rus.fit_resample(data, label)
undersampled_data = pd.DataFrame(undersampled_data, columns=data.columns)
```

```python
print('원본 데이터의 클래스 비율 \n{0}'.format(pd.get_dummies(label).sum())) #데이터를 dummies에 가져와서 그 안에서 합계구하기

#숫자가 제일 큰 M이 가장 큰 영향력을 가진다. 숫자가 부족한 클래스를 숫자를 늘려 균형을 맞춘다.  
print('.\nRandom Over 샘플링결과 \n{0}'.format(pd.get_dummies(oversampled_label).sum()))

print('.\nRandom Under 샘플링결과 \n{0}'.format(pd.get_dummies(undersampled_label).sum()))
```

```python
# SMOTE

from imblearn.over_sampling import SMOTE
smote = SMOTE(k_neighbors=5)            #이웃이 5
```

```python
from sklearn.datasets import make_classification
data, label = make_classification(n_samples=1000,          #표본데이터수
                                  n_features=2,            #독립변수 수 x,y 값
                                  n_informative=2,
                                  n_redundant=0,
                                  n_repeated=0,            #중복된 성분 수
                                  n_classes=3,             #class의 수
                                  n_clusters_per_class=1,
                                  weights=[0.05, 0.15, 0.8], #weights는 가중치
                                  class_sep=0.8,
                                  random_state=2019) 
```

```python
label
```

```python
fig = plt.Figure(figsize=(12,6))
plt.scatter(data[:,0],data[:,1],c=label,linewidth=1,edgecolor='black')  #x축, y축

```

```python
smoted_data, smoted_label = smote.fit_resample(data, label)
```

```python
print('원본 데이터의 클래스 비율\n{0}'.format(pd.get_dummies(label).sum()))
print('Smoted 데이터의 클래스 비율\n{0}'.format(pd.get_dummies(smoted_label).sum()))
```

```python
fig = plt.Figure(figsize=(12,6))
plt.scatter(smoted_data[:,0],smoted_data[:,1], c=smoted_label, linewidth=1, edgecolor = 'black')  
#노란색은 거의 변화없음. 빈자리에 나머지 초록색, 검은색이 증가함(검은색 = 0, 초록색 = 1, 노란색 =2)
```

```python
#차원의 축소(차원의 저주)   #나만의 tip) 3차원에서 2차원으로 볼때, 위에서 본다.
#차원의 축소의 이유 : 데이터의 밀도를 높여서 학습의 효율을 높이는 것

#차원의 축소
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
print(digits.DESCR)  #DESCR 어떤 데이터인지 설명
```

```python
data = digits.data
label = digits.target
```

```python
print(data[0])   #64차원, 숫자 1개를 나타나기 위해 64픽셀 필요. 내용은 별로 없어서서 밀도가 낮다. 그래서서 차원의 저주가 일어남.
print(label[0])  #0의 실제값
```

```python
#데이터의 형태를 reshape()을 사용해서 바꿀 수 있다. 
#print(data[0].reshape((8,8)))  #모양 변경 64개를 8*8, 여전히 64차원
print(data[0].reshape((8,-1)))  # reshape(8,8)을 좀더 쉽게 표현현 
                                #-1 자리는 너가 알아서 개개수정해!! 라는 뜻, 8개씩 끊어서 남는 개수 만큼 만들어주세요
```

```python
plt.imshow(data[0].reshape((8,8)))  #튜플 타입이어서 () 안에 쓰기 (8,8)
print('Label : {}'.format(label[0]))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)        #PCA = 주성분 분석(Principal Component Analysis)
```

```python
#데이터에서 주 성분을 찾기            # tip)축소된 차원에서 복구 불가능. 굳이 그럴 필요없다.  
new_data = pca.fit_transform(data)
```

```python
# 결과 살펴보기
print('원본 데이터의 차원 \n{}'.format(data.shape))
print('PCA 데이터의 차원 \n{}'.format(new_data.shape))
```

```python
plt.scatter(new_data[:,0],new_data[:,1], c= label, linewidth=1, edgecolor='black')  
#데이터들이 집합을 유지하고 있음. 차원을 줄여 놓고 특징 살피기기
#데이터 밀도 높이기
#참고: 이 방법으로 PCA를 3으로 잡을 때, 3차원으로 그려지지 않는다. 3차원 그리는 방법은 따로 있다.
```
